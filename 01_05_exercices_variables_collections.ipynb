#### Exercice 3


Write a Python program to assign notes to students at the end of the year. The program must perform the following operations:

    - (a) Request a student number.
    - b) Request the student's practical note.
    - c) Request the student's test note.
    - d) Calculate if the student's average note. If the average note ( practice and test ) is less than 40, the student should automatically receive a grade of F, and the program should display the grade and exit without performing the following steps.
    - e) Request the student's exam score.
    - f) Calculate the student's final score. The practice and test scores count for 25% of the final grade each, and the final exam should count for the remaining 50%.
    - g) Calculate and display the student's score and endorsement, according to the following table:

| final note  | classification |
| ----------- | ----- |
| 80 <= note <= 100 | A |
| 70 <= note < 80 | B |
| 60 <= note < 70 | C |
| 50 <= note < 60 | D |
| note < 50 | E |
