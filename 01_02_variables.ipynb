### The scope of variables

Not all variables are accessible from all parts of our program, and not all variables exist for the same amount of time. Where a variable is accessible and how long it exists depends on how it is defined. We call the part of a program where a variable is accessible its scope, and the duration for which the variable exists:

```python
# This is a global variable
a = 0

if a == 0:
    #  This is also a global variable
    b = 1

def my_function(c):
    # d is a local variable
    d = 3
    print(c)
    print(d)

# Call a function, with parameter 7
my_function(7)
# a and b still exist
print(a)
print(b)

# c and d don't exist anymore -- these statements will give us name errors!
print(c)
print(d)
```

### The `=` operator

This operator assigns the value on the right side to the variable on the left side, sometimes creating the variable first. If the right side is an expression (such as an arithmetic expression), it will be evaluated before the assignment occurs. Here are some examples:

```python
total = 3

a_number = 5 # assign the value 5 to the variable a_number
a_number = total # the variable a_number takes the value of the variable total
a_number = total + 5 # a_number takes the value of total + 5
a_number = a_number + 1 # a_number becomes a_number + 1


# Illegal statements:
3 = 4
3 = a
a + b = 3

# Multiple assignment: variables a and b will take the value 0
a = b = 0

# Illegal declaration:
a = 0 = b
```

## 2. Conversions

As we write programs, we will often find that we need to convert data from one type to another, for example from a string to an integer or from an integer to a floating point number. There are two types of type conversions in Python: 
    
    implicit and explicit conversions.

### Implicit conserves

Recall from the section on floating-point operators that we can arbitrarily combine integers and floating-point numbers in an arithmetic expression - and that the result of such an expression will always be a floating-point number. This is because Python will convert the integers to floating point numbers before evaluating the expression. This is an implicit conversion - we don't have to convert anything ourselves. There is usually no loss of precision when an integer is converted to a floating point number.



```python
a = 8.5 * 2
print(a)

b = 3 / 2
print(b)

c = 8.5 + 7 // 3 - 2.5
print(c)
```

### The explicit conserves (cast)


Converting numbers from `float' to `int' will result in a loss of precision. For example, try to convert `5.834` to `int`: it is not possible to do so without losing precision. For this to happen, we must explicitly tell Python that we are aware that precision will be lost.
   
For example, we must tell the compiler to convert a float to an int like this:


```python
i = int(5.834)
print(i)
```

The function `int` converts a float to an int by removing the decimal part: it will always be rounded down.
