## 2. Variables
When we write a Python program, we will create variables that store values such as numbers or strings, as well as functions and classes. These variables must have a unique name that respects the following rules:
  - It can only contain letters (upper or lower case), numbers or the underscore character (_) (no spaces!).
  - It cannot start with a number.
  - It is not a keyword.

## 6. The integers (int)

An integer (of type `int`) is a whole number such as 1, 5, 1350 or -34. Numbers with decimal points are float numbers. 2.3, 12.34, 1.0 are not integers but floats.

We can perform the following operations on integers :

| Operation | Symbol | Example | Result |
| --------- | ------ | ------- | ------ |
| Addition | + | 28 + 10 | 38 |
| Subtraction | - | 28 - 10 | 18 |
| Multiplication | * | 28 * 10 | 280 |
| Division // | 28 // 10 | 2 |
| Modulo (rest) | % | 28 % 10 | 8 |
| Power (power) | ** | 28**10 | 296196766695424 |

````python
print(3)

# We can also do operations on integers
print(1 + 2)
print (15 + 20 * 3)

example = 13 // 2 + 3
print (example)
````

## 7. Floating point numbers (float)

Float numbers are numbers with a decimal point or an exponent (or both). Examples are `5.0`, `10.24`, `0.0`, `12.` and `0.3`.

We can use scientific notation to denote very large or very small floating point numbers, for example 3.8 x 10^15 would be written as `3.8e15` or `3.8e+15`.

### Division

The arithmetic operators for `float` are the same as for integers. Except for division - the floating point division operator is `/`. Float operations always produce a floating point solution. The order of precedence for these operators is the same as for integer operators.

````python
a = 1.5 + 2
print(type(a))
print("a = ", a )

b = 1.5 // 2.0 
print(type(b))
print("b = ", b )

c = 1.5 / 2.0 
print(type(c))
print("c = ", c )

d = 1 / 2
print(type(d))
print("d = ", d )

e = -3 // 2
print(type(e))
print("e = ", e )
````

## 8. Indentation

Many languages organize code into blocks using braces (`{` and `}`) or `BEGIN` and `END` statements, these languages encourage us to indent blocks to make the code easier to read, but indentation is not mandatory. Python uses indentation only to delimit blocks, so we must indent our code:

````python
# Indentation is required when you create some python entities that define a code block that will be executer when the entity is triggered
# Examples :

# Functions
def my_fuction():
    print("Hello World in 'function'")
# function execution is outside function block
my_fuction()

# If statements
if "Hello" != "World":
    print("Hello World in 'if'")
    
# Loops
for i in [0, 1, 2]:
    print(f"Hello World in 'loop {i}'")
````
