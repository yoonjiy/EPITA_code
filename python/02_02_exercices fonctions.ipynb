## Exercice 3
 - 1) Write a function called `calculator`. It must take the following parameters: two numbers, an arithmetic operation (which can be addition, subtraction, multiplication or division and is addition by default), and an output format (which can be integer or floating point, and is floating point by default). The division must be a floating point division.
 
  The function must perform the requested operation on the two input numbers and return a result 

 - 2) Call the function with the following sets of parameters and check that the response is what you expect:
     - `2`, `3.0`
     - 2, `3.0`, the operation is division
     - 2, `3.0`, operation is division, output format is integer

## Exercice 4

Rewrite the `calculator` function from the previous exercise so that it takes a *some number of numbers* and the same optional parameter( or parameters (you think about it)). The function should apply the operation to the first two numbers, then apply it again to the result and the next number, and so on. For example, if the numbers are 6, 4, 9 and 1 and the operation is subtraction, the function should return `6 - 4 - 9 - 1`. If only one number is entered, it should be returned as it is. If no number is entered, raise an exception.
